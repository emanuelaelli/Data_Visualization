```python
valore1 = ['Utilità']*29
valore2 = ['Chiarezza']*29
valore3 = ['Informatività']*29
valore4 = ['Bellezza']*29
valore5 = ['Valore complessivo']*29
```

# Inizio visualizzazioni

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Visualizzazione 1

```python
df1 = pd.DataFrame(list(zip(valore1, Utilità)), columns = ['Valore valutato', 'Voto'])
df2 = pd.DataFrame(list(zip(valore2, Chiarezza)), columns = ['Valore valutato', 'Voto'])
df3 = pd.DataFrame(list(zip(valore3, Informatività)), columns = ['Valore valutato', 'Voto'])
df4 = pd.DataFrame(list(zip(valore4, Bellezza)), columns = ['Valore valutato', 'Voto'])
df5 = pd.DataFrame(list(zip(valore5, Valore_compl)), columns = ['Valore valutato', 'Voto'])
df1_2 = df1.merge(df2, how='outer')
df1_2_3 = df1_2.merge(df3, how='outer')
df1_2_3_4 = df1_2_3.merge(df4, how='outer')
df = df1_2_3_4.merge(df5, how='outer')
df['Tipo votazione'] = (['Ciao']*145)
df.loc[df.Voto<=3,'Tipo votazione']='1-3'
df.loc[df.Voto>3,'Tipo votazione']='4-6'
ax = sns.histplot(df, y="Valore valutato", hue='Tipo votazione', stat='percent', multiple='stack', shrink=.7, palette=['blue', 'red'])
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel('Percentuale')
plt.ylabel('')
plt.savefig('Questionario1.png', dpi=300, facecolor='w', bbox_inches='tight', pad_inches=0)
```
